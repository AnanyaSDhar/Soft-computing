```python
# Assuming you have already defined mclusters and pclusters

# Function to calculate Jaccard similarity
def jaccard_similarity(list1, list2):
    intersection = len(set(list1).intersection(list2))
    union = len(set(list1).union(list2))
    return intersection / union if union != 0 else 0

# Create a similarity matrix C
p = len(pclusters)
C = np.zeros((p, p))

for i in range(p):
    for j in range(p):
        C[i, j] = jaccard_similarity(pclusters[i], pclusters[j])

print("Similarity Matrix C:")
for i in range(p):
    for j in range(p):
        print(C[i,j], end =", ")
    print("\n")
```

```python
# Find the indices of the maximum value(s) in the similarity matrix
max_indices = np.unravel_index(np.argmax(C), C.shape)
max_i, max_j = max_indices

# Merge the most similar clusters
new_cluster = pclusters[max_i] + pclusters[max_j]

# Remove the merged clusters from pclusters
pclusters.pop(max_i)
if max_i < max_j:
    pclusters.pop(max_j - 1)
else:
    pclusters.pop(max_j)

# Append the new merged cluster
pclusters.append(new_cluster)

print("Merged Cluster:")
print(new_cluster)
print("Updated pclusters:")
print(pclusters)
```
